```python
import pandas as pd 
import re

data = pd.read_excel('/Users/dianashevtsova/Desktop/R/python datasets/data.xlsx')

def month(status):
    matching_month = re.search(r'(\w+)\s+\d{4}', str(status))
    if matching_month:
        exact_month = matching_month.group(1)
        return exact_month
    return None

data['month'] = data['status'].apply(month)
data['month'].fillna(method='ffill', inplace=True)

data

```

```python
from datetime import datetime

def calculation_bonus(data, cutoff_date_str='2021-07-01'):
    cutoff_date = datetime.strptime(cutoff_date_str, '%Y-%m-%d')
    
    if data['new/current'] == 'новая' and data['status'] == 'ОПЛАЧЕНО' and data['document'] == 'оригинал':
        if data['receiving_date'] <= cutoff_date:
            return data['sum'] * 0.07  
        else:
            return 0  

   
    elif data['new/current'] == 'текущая' and data['status'] != 'ПРОСРОЧЕНО' and data['document'] == 'оригинал':
        if data['receiving_date'] <= cutoff_date:
            if data['sum'] > 10000:
                return data['sum'] * 0.05  
            else:
                return data['sum'] * 0.03  
        else:
            return 0  
    
    return 0  

data['bonus'] = data.apply(lambda x: calculation_bonus(x, cutoff_date_str='2021-07-01'), axis=1)

# Функция для расчёта остатков
def calculate_remaining(data, cutoff_date_str='2021-07-01'):
    cutoff_date = datetime.strptime(cutoff_date_str, '%Y-%m-%d')
    
    if data['new/current'] == 'новая' and data['status'] == 'ОПЛАЧЕНО' and data['document'] != 'оригинал' and data['receiving_date'] > cutoff_date:
        return data['sum'] * 0.07
    elif data['new/current'] == 'текущая' and data['status'] != 'ПРОСРОЧЕНО' and data['document'] != 'оригинал' and data['receiving_date'] > cutoff_date:
        if data['sum'] > 10000:
            return data['sum'] * 0.05
        else:
            return data['sum'] * 0.03
    return 0

data['remaining'] = data.apply(lambda x: calculate_remaining(x, cutoff_date_str='2021-07-01'), axis=1)

data
```
